Let's assume that the choice of the next action occurs equally likely in the interval -1,1. In addition, each time we add normally distributed noise $N~(0,\sigma^2)$. The form of its probability density function is

$$
f(x) = \int_{-1}^{1} \frac{1}{\sigma \sqrt{2 \pi}} \exp{-\frac{(x-\mu)^2}{2 \sigma^2}}\,d \mu 
$$

$$
f(x) = \Phi\left(\frac{1 - x}{\sigma}\right) - \Phi\left(-\frac{1 + x}{\sigma}\right)
$$

where $\Phi(z)$ is the cumulative distribution function of the standard normal distribution.

Thus, adding noise greatly increases the probability of falling outside the edges of the interval -1,1.
